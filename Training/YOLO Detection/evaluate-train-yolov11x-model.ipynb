```python
from ultralytics import YOLO

# Load trained YOLOv11x model
model = YOLO("/kaggle/input/yolov11-skinz/pytorch/default/1/best-v11.pt")

# Run validation using your data.yaml
metrics = model.val(data="/kaggle/input/skin-lesion-detection-dataset/data.yaml")

# Print evaluation results (no parentheses!)
print(f"mAP@0.5: {metrics.box.map50:.2f}%")
print(f"mAP@0.5:0.95: {metrics.box.map:.2f}%")
print(f"Precision: {metrics.box.mp:.2f}%")
print(f"Recall: {metrics.box.mr:.2f}%")
```
